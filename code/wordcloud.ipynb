```python
mask_image = np.array(Image.open('electric-guitar.jpg'))
```

```python
text = "el otro día fui a tu casa a ver una pelicula de vaqueros no me gusto todos morían prefiero cuando al menos uno sobrevive también me gusta cuando te tumbas al sol y la luz del verano se refleja en tu piel recién salida del mar de otoño contigo se para el tiempo no me importa cuando empezó el universo si este se expande o si por el contrario a las ultimas tendencias filosoficas vivimos en una simulacion yo vivo en la tibieza de tu piel al sol finalmente estos esfuerzos literarios no llevan a ninguna parte más que a ser una expresión modesta de deseo y mediterráneo"
```

```python
# mask[150:350,150:350,:] = 255 #masked out area
```

```python
# mask2 = np.array(Image.open('my_image.jpg')) #to apply own black/white mask
```

```python
cloud = wordcloud.WordCloud(background_color='white',
                           max_words=50,
                           mask=mask_image,
                           collocations=False, #calculates frequencies
                           contour_color='steelblue').generate(text)
```

```python
plt.figure(figsize=(40,40))
plt.imshow(cloud, interpolation='bilinear')
plt.savefig('cloud.png')
plt.show()
```
