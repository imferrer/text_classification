```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
import wordcloud
```

```python
from PIL import Image
```

```python
text = 'jglekjrg elkfj  erkl g'
```

```python
mask = np.zeros((500,500,3),np.uint8)
```

```python
mask[150:350,150:350,:] = 255 #masked out area
```

```python
# mask2 = np.array(Image.open('my_image.jpg')) #to apply own black/white mask
```

```python
cloud = wordcloud.WordCloud(background_color='white',
                           max_words=50,
                           mask=mask,
                           collocations=False, #calculates frequencies
                           contour_color='steelblue').generate(text)
```

```python
plt.figure(figsize=(40,40))
plt.imshow(cloud, interpolation='bilinear')
plt.savefig('cloud.png')
plt.show()
```
